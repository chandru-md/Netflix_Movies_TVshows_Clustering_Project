```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sentence_transformers import SentenceTransformer
from sklearn.decomposition import TruncatedSVD
```

```python
tv_data["combined_text"] = (
    tv_data["description"] + " " + tv_data["listed_in"]
)
```

```python
tfidf = TfidfVectorizer(
    stop_words="english",
    max_features=10000,
    ngram_range=(1,2)
)

X_tv = tfidf.fit_transform(tv_data["combined_text"])
```

```python
svd = TruncatedSVD(n_components=100, random_state=42)
X_tv_reduced = svd.fit_transform(X_tv)
```

```python
for k in range(2, 12):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_tv_reduced)
    score = silhouette_score(X_tv_reduced, labels)
    print(f"K={k}, Score={score}")
```

```python
kmeans = KMeans(n_clusters=10, random_state=42)
tv_data["cluster"] = kmeans.fit_predict(X_tv_reduced)
```

```python
for i in range(tv_data["cluster"].nunique()):
    print(f"\nCluster {i}")
    print(tv_data[tv_data["cluster"] == i]["listed_in"].value_counts().head(5))
```
